```python
import pandas as pd
from pymongo import MongoClient
```

```python
df = pd.read_csv("USDA.csv")
df.head()
```

```python
df = df[['Description', 'Calories', 'Protein', 'TotalFat', 'Carbohydrate', 'Cholesterol', 'Sugar', 'Calcium', 'Iron', 'Potassium']]
df.dropna()
df.head()
```

```python
df.rename(columns = {
    'Description': 'meal_name',
    'Calories': 'calories',
    'Protein': 'proteins',
    'TotalFat': 'fats',
    'Carbohydrate': 'carbs',
    'Cholesterol': 'cholestrol',
    'Sugar': 'sugar',
    'Calcium': 'calcium',
    'Iron': 'iron',
    'Potassium': 'potassium'
}, inplace = True)
df.head()
```

```python
data = df.to_dict(orient = 'records')
data
```

```python
client = MongoClient("mongodb://localhost:27017")
db = client["meal_planner_db"]
collection = db["foods"]
collection
```

```python
collection.insert_many(data)
```
